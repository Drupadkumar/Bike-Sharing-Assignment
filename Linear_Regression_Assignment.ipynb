```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Generic function to calculate VIF of variables
def calculate_vif(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
#calculate with RFE selected variables
X_train_rfe = X_train[col]
```

```python
#calculate VIF
calculate_vif(X_train_rfe)
```

Humidity has high VIF

#5. Building a linear model

```python
# Building first linear model
X_train_lm_1 = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(y_train, X_train_lm_1).fit()
print(lr_1.summary())
```

```python
# As humidity shows high VIF, drop Humidity
X_train_new = X_train_rfe.drop(['humidity'], axis=1)
```

```python
calculate_vif(X_train_new)
```

VIF seems good now. Lets see if we can reduce further

```python
X_train_lm_2= sm.add_constant(X_train_new)
lr2 = sm.OLS(y_train, X_train_lm_2).fit()
print(lr2.summary())
```

```python
# drop nov variable, since nov is having high p-value
X_train_new = X_train_new.drop(['nov'], axis=1)
calculate_vif(X_train_new)
```

```python
# Building 3rd LR
X_train_lm_3= sm.add_constant(X_train_new)
lr3 = sm.OLS(y_train, X_train_lm_3).fit()
print(lr3.summary())
```

```python
X_train_new = X_train_new.drop(['dec'], axis=1)
calculate_vif(X_train_new)
```

```python
# Building 4th LR
X_train_lm_4= sm.add_constant(X_train_new)
lr4 = sm.OLS(y_train, X_train_lm_4).fit()
print(lr4.summary())
```

```python
X_train_new = X_train_new.drop(['jan'], axis=1)
calculate_vif(X_train_new)
```

```python
#Building 5th LR
X_train_lm_5= sm.add_constant(X_train_new)
lr5 = sm.OLS(y_train, X_train_lm_5).fit()
print(lr5.summary())
```

```python
X_train_new = X_train_new.drop(['jul'], axis=1)
calculate_vif(X_train_new)
```

```python
# Building 6th LR
X_train_lm_6= sm.add_constant(X_train_new)
lr6 = sm.OLS(y_train, X_train_lm_6).fit()
print(lr6.summary())
```

We can cosider the above model i.e lr6, as it seems to have very low multicolinearity between the predictors and the p-values for all the predictors seems to be significant.
F-Statistics value of 248.4 (which is greater than 1) and the p-value of 1.47e-186 i.e almost equals to zero, states that the overall model is significant
